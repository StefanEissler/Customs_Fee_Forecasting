```python
import pandas as pd 

customer_a_matrix = pd.read_csv("../data/evaluation/customer_a_evaluation.csv")
customer_b_matrix = pd.read_csv("../data/evaluation/customer_b_evaluation.csv")
customer_c_matrix = pd.read_csv("../data/evaluation/customer_c_evaluation.csv")
customer_d_matrix = pd.read_csv("../data/evaluation/customer_d_evaluation.csv")
customer_e_matrix = pd.read_csv("../data/evaluation/customer_e_evaluation.csv")
customer_f_matrix = pd.read_csv("../data/evaluation/customer_f_evaluation.csv")
customer_g_matrix = pd.read_csv("../data/evaluation/customer_g_evaluation.csv")
customer_h_matrix = pd.read_csv("../data/evaluation/customer_h_evaluation.csv")
customer_i_matrix = pd.read_csv("../data/evaluation/customer_i_evaluation.csv")

customer_a_matrix['customer'] = 'a'
customer_b_matrix['customer'] = 'b'
customer_c_matrix['customer'] = 'c'
customer_d_matrix['customer'] = 'd'
customer_e_matrix['customer'] = 'e'
customer_f_matrix['customer'] = 'f'
customer_g_matrix['customer'] = 'g'
customer_h_matrix['customer'] = 'h'
customer_i_matrix['customer'] = 'i'

all_customers = pd.concat([customer_a_matrix, customer_b_matrix, customer_c_matrix,
                           customer_d_matrix, customer_e_matrix, customer_f_matrix,
                           customer_g_matrix, customer_h_matrix, customer_i_matrix])

all_customers.reset_index(drop=True, inplace=True)
all_customers.to_csv("../data/evaluation/all_customers.csv", sep=';', index=False)

all_customers
```
